```python
from k_means import K_Means_Classifier
import random
import matplotlib.pyplot as plt
```

```python
def generate_data():
    cluster1 = [random.random()*2 for _ in range(100)]
    cluster1_y = [random.random()*2+.5 for _ in range(100)]
    cluster2 = [random.random()*2+3.5 for _ in range(100)]
    cluster2_y = [random.random()*2+7.5 for _ in range(100)]
    cluster3 = [random.random()*2+7.5 for _ in range(100)]
    cluster3_y = [random.random()*2+.5 for _ in range(100)]
    x = cluster1 + cluster2 + cluster3
    y = cluster1_y + cluster2_y + cluster3_y
    return x, y
X, y = generate_data()
plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.scatter(X,y)
```

```python
clf = K_Means_Classifier()
X_train = [[a, b] for a,b in zip(X,y)]
clusters = clf.fit(X_train)
centroids = clf.centroids
centroids_x = [c[0] for c in centroids]
centroids_y = [c[1] for c in centroids]

color_map = []
colors = ["red", "green", "blue"]
for i in range(len(X_train)):
    for j, c in enumerate(clusters):
        if i in c:
            color_map.append(colors[j])

plt.xlim(-5, 10)
plt.ylim(-5, 10)
plt.scatter(X, y, c=color_map)
plt.scatter(centroids_x, centroids_y, color="black")
```
